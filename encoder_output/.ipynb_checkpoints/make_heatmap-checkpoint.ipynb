```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
tmp = torch.load('./10x12x10x128x768.pt',map_location=torch.device('cpu'))
```

[N/batch, layer_num, batch, sequence_length, token_dimension]

```python
for batch in range(0,10):
    a = tmp[0]
    sum_ = 0
    tmp_result = []
    real_result = []
    for layer_num in range(0, 12):
        sum_ = 0
        #print(i,'번째 레이어\n', a[i][z])
        for token_num in range(0, 768):
            for seq_len in range(0, 128):
                sum_ = sum_ + (a[layer_num][batch][seq_len][token_num])
            if (token_num+1)%64 == 0:
                tmp_result.append(abs(sum_)) # 토큰에 128x64의 합..
                                             # 0번째 레이어의 12개 토큰이 tmp_result에 저장됨.
        real_result.append(tmp_result)
        # 최종적으로 real_result에는 12개의 레이어 x 12개의 토큰이 저장됨.
    a = list(map(lambda x: x.numpy(), real_result[0]))

    m1 = []
    m2 = []
    for cnt, _ in enumerate(a):
        m1.append(_)
        if (cnt+1)%12 == 0:
            m2.append(m1)
            m1=[]

    plt.figure(figsize=(15,15))
    sns.heatmap(m2, cmap='viridis', annot=True, vmin=0, vmax=7000) #숫자 출력
    plt.savefig('./image/image'+str(batch)+'.png', format='png', dpi=300)
    #plt.show()
    # 행 -> layer
    # 열 -> Head
```

```python
"""
concat 이후에 헤드별로 결과값을 나눠서 히트맵 그리기
elements[layer][batch][seq][token]
"""
import torch
import matplotlib.pyplot as plt
import seaborn as sns
tmp = torch.load('./test.pt',map_location=torch.device('cpu'))
tmp = tmp[:12] # layer num
Heads = []
Head = []
batch_result = []
sum_ = 0
for batch_size in range(0, 10):
    for layer_num in range(0, 12):
        for token_num in range(0, 768):
            for seq_num in range(0, 128):
                #print(tmp[layer_num][batch_size][seq_num][token_num])
                sum_ = sum_ + tmp[layer_num][batch_size][seq_num][token_num]
            if (token_num+1)%64 == 0:
                sum_ = sum_.numpy()
                Head.append(sum_)
                sum_ = 0
        Heads.append(Head)
        Head=[]
    batch_result.append(Heads)
    Heads=[]
    
m1 = []
m2 = []
for cnt, batch_datas in enumerate(batch_result):
    for cnt2, layers_data in enumerate(batch_datas):
        m1.append(layers_data)
        m2 = m2 + layers_data
    plt.figure(figsize=(15,15))
    sns.heatmap(m1, cmap='viridis', annot=True, vmin = min(m2), vmax = max(m2)) #숫자 출력
    plt.savefig('./2. image_normal/'+str(cnt)+'-image.jpeg', format='jpeg', dpi=300)
    #plt.show()
    m1 = []
    m2 = []
```

```python
"""
concat 이후에 헤드별로 결과값을 나눠서 히트맵 그리기(절대값)
elements[layer][batch][seq][token]
"""
import torch
import matplotlib.pyplot as plt
import seaborn as sns
tmp = torch.load('./test.pt',map_location=torch.device('cpu'))
tmp = tmp[:12] # layer num
Heads = []
Head = []
batch_result = []
sum_ = 0
for batch_size in range(0, 10):
    for layer_num in range(0, 12):
        for token_num in range(0, 768):
            for seq_num in range(0, 128):
                #print(tmp[layer_num][batch_size][seq_num][token_num])
                sum_ = sum_ + tmp[layer_num][batch_size][seq_num][token_num]
            if (token_num+1)%64 == 0:
                sum_ = sum_.numpy()
                Head.append(abs(sum_))
                sum_ = 0
        Heads.append(Head)
        Head=[]
    batch_result.append(Heads)
    Heads=[]
    
m1 = []
m2 = []
for cnt, batch_datas in enumerate(batch_result):
    for cnt2, layers_data in enumerate(batch_datas):
        m1.append(layers_data)
        m2 = m2 + layers_data
    plt.figure(figsize=(15,15))
    sns.heatmap(m1, cmap='viridis', annot=True, vmin = 0, vmax = 400) #숫자 출력
    plt.savefig('./3. image_abs/'+str(cnt)+'-image.png', format='png', dpi=300)
    #plt.show()
    m1 = []
    m2 = []
```

```python
"""
concat이후 헤드별로 결과값을 나눠서
각 인코더별 input과 output의 L2-Norm 그리기
elements[layer][batch][seq][token]
"""
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

tmp = torch.load('./test.pt',map_location=torch.device('cpu'))
tmp = tmp[:12] # layer num
Heads = []
Head = []
batch_result = []
sum_ = 0
for batch_size in range(0, 10):
    for layer_num in range(0, 12):
        for token_num in range(0, 768):
            for seq_num in range(0, 128):
                #print(tmp[layer_num][batch_size][seq_num][token_num])
                sum_ = sum_ + tmp[layer_num][batch_size][seq_num][token_num]
            if (token_num+1)%64 == 0:
                sum_ = sum_.numpy()
                Head.append(sum_)
                sum_ = 0
        Heads.append(Head)
        Head=[]
    batch_result.append(Heads)
    Heads=[]
```

```python
m1 = []
m2 = []
l2_result = []
for cnt, batch_datas in enumerate(batch_result):
    for cnt2, layers_data in enumerate(batch_datas):
        m1.append(layers_data)
        m2 = m2 + layers_data
        if len(m1) == 2:
            #print(m1)
            L2_norm = np.linalg.norm(m1, axis=0, ord=2)
            l2_result.append(L2_norm)
            m1.pop()
    plt.figure(figsize=(15,15))
    sns.heatmap(l2_result, cmap='viridis', annot=True, vmin = 0, vmax = max(m2)) #숫자 출력
    plt.savefig('./4. image_L2/'+str(cnt)+'-image.png', format='png', dpi=300)
    #plt.show()
    m1 = []
    m2 = []
    l2_result = []
```
